```python
# test 2
import turtle  
 
t = turtle.Turtle()  
s = turtle.Screen()  
s.bgcolor("black")  
t.pencolor("red")  
  
a = 0  
b = 0  
t.speed(0)  
t.penup()  
t.goto(0,200)  
t.pendown()  
while(True):  
    t.forward(a)  
    t.right(b)  
    a+=3  
    b+=1  
    if b == 210:  
        break  
    t.hideturtle()  
  
turtle.done()  
```

```python
# test 4
import turtle 

t = turtle.Turtle() #Turtle object
wn = turtle.Screen() #Screen Object
wn.bgcolor("black") #Screen Bg color
wn.title("Fractal Tree Pattern")
t.left(90) #moving the turtle 90 degrees towards left
t.speed(10)#setting the speed of the turtle


def  draw(distance=20,size=2,color="yellow",draw_factor=3/4,left_value=30,right_value=60): #recursive function taking length 'l' as argument
    if(distance<10):
        return
    else:
        t.pensize(size) #Setting Pensize
        t.pencolor(color) #Setting Pencolor as yellow
        t.forward(distance) #moving turtle forward by 'l'
        t.left(left_value) #moving the turtle 30 degrees towards left
        draw(distance*draw_factor,color=color) #drawing a fractal on the left of the turtle object 'roo' with 3/4th of its length
        t.right(right_value) #moving the turtle 60 degrees towards right
        draw(distance*draw_factor,color=color) #drawing a fractal on the right of the turtle object 'roo' with 3/4th of its length
        t.left(left_value) #moving the turtle 30 degrees towards left
        t.pensize(size)
        t.backward(distance) #returning the turtle back to its original psition

draw(distance=20,size=2,color="yellow",draw_factor=3/4) # drawing 20 times 

t.right(90)

draw(color="magenta")

t.left(270)

draw(color="red")

t.right(90)

draw(color="white")

# ########################################################

draw (distance=40,size=3,color="lightgreen",draw_factor=4/5)

t.right(90)

draw (distance=40,size=3,color="red",draw_factor=4/5)

t.left(270)

draw (distance=40,size=3,color="YELLOW",draw_factor=4/5)

t.right(90)

draw (distance=40,size=3,color="white",draw_factor=4/5)

# ########################################################
        
draw (distance=60,size=2,color="cyan",draw_factor=6/7)

t.right(90)
        
draw (distance=60,size=2,color="yellow",draw_factor=6/7)

t.right(90)
        
draw (distance=60,size=2,color="magenta",draw_factor=6/7)

t.right(90)

draw(distance=60,size=2,color="white",draw_factor=6/7)
# wn.exitonclick()



turtle.done() 
```
